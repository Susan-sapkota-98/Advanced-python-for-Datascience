```python
def myDecorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper
```

```python
@myDecorator
def sayHello():
    print("Hello")
# sayHello=myDecorator(sayHello)
# sayHelloo()
```

```python
import time
def timer(func):
    def wrapper(*args,**kwargs):
        start=time.time()
        result=func(*args,**kwargs)
        end=time.time()
        print(f"Elapsed time={end-start} seconds")
        return result
    return wrapper
```

```python
@timer
def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*i
    return f
fact(100)
```

```python
def squares(n=10):
    print(f'Genreating the squares from 1 to {n ** 2}')
    for i in range(1,n+1):
        yield i **2
gen=squares()
gen
```
