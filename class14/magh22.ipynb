```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# y = m*X + c
X = np.arange(0,30,0.277)
print(X[:3])
y = 0.9912*X + 0.7734
print(y[:3])
print(len(X), len(y))
```

```python
X_mean=np.mean(X)
y_mean=np.mean(y)
print(X_mean)
print(y_mean)
```

```python
n=len(X)
cov_x_y=(1/n)*np.sum((X-X_mean)*(y-y_mean))
print(cov_x_y)
cov_x=(1/n)*np.sum(np.square(X-X_mean))
print(cov_x)
```

```python
b1 = cov_x_y/cov_x
b0 = y_mean - b1*X_mean
print(f'y = {b0:.4f} + {b1:.4f}*X')
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
X = np.arange(0,10,0.277)
y = 0.9912*X + 0.7734
print(len(X), len(y))
```

```python
split = int(len(X)*0.8)
X_train, y_train = X[:split], y[:split]
X_test, y_test = X[split:],y[split:]
```

```python
def plot_data(X_train=X_train,y_train=y_train,
              X_test=X_test, y_test=y_test, predictions=None):
    plt.scatter(X_train, y_train, c='b', label="Training Data")
    plt.scatter(X_test, y_test, c='r', label="Testing Data")
    if predictions is not None:
        plt.scatter(X_test, predictions, c='g', label="Predicted Data")
    plt.title('X vs y')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.show()

plot_data()
```

```python
# pip install scikit-learn
# Create a LR MODEL
lr_model = LinearRegression()
# Train the model
lr_model.fit(X_train.reshape(-1,1), y_train)
# Get the predictions
predictions = lr_model.predict(X_test.reshape(-1,1))
```

```python
plot_data(predictions=predictions)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_true=y_test, y_pred=predictions))
```
