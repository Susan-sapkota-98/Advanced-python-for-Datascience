```python
#to calculate the frequency of the words
aDict={}
with open(file='test.txt',mode='r') as f:
    content=f.read().lower()
    words=content.split()
    for word in words:
        if word not in aDict:
            aDict[word]=1
        else:
            aDict[word]+=1
print(aDict)
```
