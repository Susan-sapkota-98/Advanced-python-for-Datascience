```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# NumPy: create data
x = np.array([1, 2, 3, 4, 5])
y = x * 2

# Pandas: create DataFrame
data = pd.DataFrame({
    "Numbers": x,
    "Doubles": y
})

print(data)

# Matplotlib: plot data
plt.plot(x, y)
plt.xlabel("Numbers")
plt.ylabel("Doubles")
plt.title("Simple Plot Example")
plt.show()
```
