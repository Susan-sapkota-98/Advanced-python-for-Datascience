```python
import numpy as np
import pandas as pd
x=np.array([10,20,30,40])
y=np.array([15,25,35,45])
np.cov(x,y)
```

```python
x=np.array([0,1,2])
y=np.array([2,1,0])
np.cov(x,y)
```

```python
df=pd.DataFrame({'X':x,'Y':y})
df.cov()
```

```python
import numpy as np
import pandas as pd
x=np.array([10,20,30,40])
y=np.array([15,25,35,45])
np.corrcoef(x,y)
```

```python
x=np.array([0,1,2])
y=np.array([2,1,0])
np.corrcoef(x,y)
```

```python
df=pd.DataFrame({'X':x,'Y':y})
df.corr()
```
