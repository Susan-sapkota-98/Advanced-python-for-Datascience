```python
from tqdm import tqdm
import random

names = ["Ram", "Hari", "shyam", "krishna", "sita"]
final_list = []

for i in tqdm(range(1, 10_000_001)):
    random_name = random.choice(names)
    final_list.append({
        "EID": i,
        "Name": random_name,
        "Salary": 100000 * random.random()
    })
```

```python
import pandas as pd
first_chunk=True
output_file='processed_data.csv'
for chunk in  pd.read_csv('big_data.csv',chunksize=100000):
    chunk["Gender"]=chunk["Name"].apply(lambda x: "Female" if x=='Sita' else 'Male')
    chunk.to_csv(
        output_file,
        index=False,
        mode="w" if first_chunk else "a",
        header=first_chunk
    )
    first_chunk=False
```
