```python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __str__(self):
        return f'({self.x}, {self.y})'

    def __add__(self,other):
        new_x=self.x+other.x
        new_y=self.y+other.y
        return Point(new_x,new_y)
p1=Point(2,3)
p2=Point(3,2) 
p3=p1+p2 #p1.__add__(p2)
print(p3)
```

```python
#Wap to create a class named Distance that has feet and inches as data member overload any 2 operators.
class Distance:
    def __init__(self, feet=0, inches=0.0):
        self.feet = feet
        self.inches = inches
        self._normalize()

    def _normalize(self):
        if self.inches >= 12:
            self.feet += int(self.inches // 12)
            self.inches = self.inches % 12

    # Addition
    def __add__(self, other):
        if isinstance(other, Distance):
            return Distance(
                self.feet + other.feet,
                self.inches + other.inches
            )
        return NotImplemented

    # Equality check
    def __eq__(self, other):
        if isinstance(other, Distance):
            return (self.feet == other.feet and
                    abs(self.inches - other.inches) < 1e-6)
        return False

    def __str__(self):
        return f"{self.feet} feet {self.inches:.2f} inches"


# Example usage
d1 = Distance(5, 9)
d2 = Distance(3, 4)
d3 = Distance(9, 1)

print("Distance 1:", d1)
print("Distance 2:", d2)
print("Addition:", d1 + d2)
print("d1 + d2 == d3 ?", d1 + d2 == d3)

```
