```python
# simple  object access protocal (SOAP)
from zeep import Client
#WSDL URL
wsdl="http://www.dneonline.com/calculator.asmx?WSDL"
client=Client(wsdl=wsdl)

#Call a SOAP method
result=client.service.Add(10,20)
print(result)
```

```python
import requests
from pprint import pprint
api_key = "1e679f6aa64bfb3ba0eabd0dff937661"
city="Dharan"
url=f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

response=requests.get(url)
if response.status_code==200:
    data=response.json()
    pprint(data)
else:
    print(response.status_code)
```

```python
import pandas as pd
from pandas import json_normalize

df = json_normalize(data)

df_weather = df[[
    'name',
    'main.temp',
    'main.humidity',
    'main.feels_like',
    'wind.speed',
    'weather'
]].copy()



df_weather['Description']=df['weather'].apply(lambda x:x[0]['description'])
df_weather=df_weather.drop(columns=['weather'])
print(df_weather)
```
