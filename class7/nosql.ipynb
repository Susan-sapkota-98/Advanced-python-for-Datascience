```python
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Select database and collection
db = client["school"]
collection = db["students"]

# Insert a single document
student = {"name": "Alice", "age": 25, "subjects": ["Math", "Science"]}
collection.insert_one(student)


```

```python
many_students=[
    {"name": "Alice", "age": 25, "subjects": ["Math", "Science"]},
    {"name": "Ram", "age": 52, "subjects": ["English", "Science"]},
]
collection.insert_many(many_students)
```

```python
#Query to display all students
for data in collection.find():
    print(data)
```

```python
#Query to display conditional results
for data in collection.find({"age":{"$gt":20}}):
    print(data)
```
