```python
#a. Reads the Excel file into a DataFrame (or using openpyxl is preferred)
from openpyxl import Workbook

wb = Workbook()
ws = wb.active

ws.append(["Name", "Math", "Science", "English"])
ws.append(["Susan", 80, 90, 40])

wb.save("students.xlsx")
```

```python
#b. Calculate the Total and Average marks for each student
from openpyxl import load_workbook

wb = load_workbook("students.xlsx")
sheet = wb.active


for row in sheet.iter_rows(min_row=2, values_only=True):
    name, math, science, english = row

    total = math + science + english
    average = total / 3

    print("Name:", name)
    print("Total:", total)
    print("Average:", average)

```

```python
from openpyxl import Workbook

new_wb = Workbook()
new_sheet = new_wb.active

new_sheet.append([
    "Name", "Math", "Science", "English",
    "Total", "Average", "Result"
])

result = "Passed" if average >= 60 else "Failed"

new_sheet.append([
    name, math, science, english,
    total, average, result
])

new_wb.save("students_result.xlsx")

print("students_result.xlsx created successfully")
```
