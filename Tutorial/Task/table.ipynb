Create a table for books with fields:title,author,year. insert 3 books and query by author delete a book

```python
import sqlite3
```

```python
conn=sqlite3.connect('books.db')
cursor=conn.cursor()
```

```python
#Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS BOOKS(
               title TEXT,
               author TEXT,
               year INTEGER
               )""")
```

```python
#Insert Data
cursor.executemany("INSERT INTO books (title,author,year) VALUES (?,?,?)",
               (["Clean Code","Robert C. Martin",2008],
                ["Harry Potter","J. K. Rowling",1997],
                ["The Hobbit","J. R. R. Tolkien",1937]
                )
               )
conn.commit()
```

```python
#Query
cursor.execute("SELECT * FROM books")
rows=cursor.fetchall()
for row in rows:
    print(row)
```

```python
#Update years
cursor.execute("UPDATE books SET year =? WHERE title=?",
               (2000,"Harry Potter") )
conn.commit()
```
