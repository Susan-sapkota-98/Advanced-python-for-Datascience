1. Write an iterator that returns the first n even numbers.

```python
class EvenNumbers:
    def __init__(self,n):
        self.n=n
        self.count=0
        self.current=0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        even=self.current
        self.current +=2
        self.count +=1
        return even

for num in EvenNumbers(5):
    print(num)
```

2. Write a generator that yields the Fibonacci sequence up to n terms.

```python
def fibo(n):
    a,b=0,1
    for _ in range(n):
        yield a
        a,b=b,a+b

for num in fibo(7):
    print(num)
```

3.  Implement a decorator that prints “Start” before a function and “End” after it.

```python
def start_end_decorator(func):
    def wrapper(*args, **kwargs):
        print("Start")
        result = func(*args, **kwargs)
        print("End")
        return result
    return wrapper

@start_end_decorator
def say_hello():
    print("Hello!")

say_hello()

```

4. Write a generator to read a CSV file line by line.

```python
def read_csv(file_path):
    with open(file_path) as f:
        for line in f:
            yield line.strip()

for line in read_csv("test.csv"):
    print(line)
```

5. Implement an iterator class that iterates over a range in reverse order

```python
class reverse:
    def __init__(self,start,end):
        self.current=end-1
        self.start=start

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current<self.start:
            raise StopIteration
        value=self.current
        self.current -=1
        return value
for num in reverse(1,5):
    print(num)
```

6. Write a generator expression to create a list of cubes of numbers from 1 to 20.

```python
def cubes(n):
    for i in range(1,n+1):
        yield i ** 3

for num in cubes(20):     
    print(num)
```

7. Create a decorator to measure memory usage of a function.

```python
import tracemalloc

def memory_usage_decorator(func):
    def wrapper(*args,**kwargs):
        tracemalloc.start()

        result=func(*args,**kwargs)

        current,peak=tracemalloc.get_traced_memory()
        tracemalloc.stop()

        print(f'Current memory usage: {current / 1024:.2f} KB')
        print(f'Peak memory usage: {peak / 1024:.2f} KB')

        return result
    return wrapper

@memory_usage_decorator
def create_list():
    data = [i for i in range(100000)]
    return data

create_list()
```

8. Write a generator that yields prime numbers indefinitely.

```python
def prime():
    num = 2
    while True:
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                break
        else:         
            yield num
        num += 1

primenum = prime()
for _ in range(10):
    print(next(primenum))

```

9. Create a decorator to cache results of a function (memoization).

```python
def memo(func):
    cache={}

    def wrapper(*args):
        if args in cache:
            print('Fetching from cache ...')
            return cache[args]
        
        print('computing results...')
        result=func(*args)
        cache[args]=result
        return result
    return wrapper
@memo
def fibo(n):
    if n <= 1:
        return n
    return fibo(n - 1) + fibo(n - 2)

print(fibo(10))
print(fibo(10))  
```

11. Write a generator to simulate a stream of sensor readings.


```python
import random
import time

def sensor_stream():
    while True:
        reading = round(random.uniform(20.0, 30.0), 2)  
        yield reading
        time.sleep(1)  
sensor = sensor_stream()

for _ in range(5):   
    print(next(sensor))
```

12. Implement a decorator to log execution time and input arguments.

```python
import time
from functools import wraps

def log_time_and_args(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()

        result = func(*args, **kwargs)

        end_time = time.time()
        execution_time = end_time - start_time

        print(f"Function: {func.__name__}")
        print(f"Arguments: args={args}, kwargs={kwargs}")
        print(f"Execution time: {execution_time:.6f} seconds")

        return result

    return wrapper

@log_time_and_args
def add(a, b):
    time.sleep(1)
    return a + b

result = add(3, 5)
print("Result:", result)

```

13. Write an iterator that yields only unique items from a list.

```python
class items:
    def __init__(self, data):
        self.data = data
        self.index = 0
        self.seen = set()

    def __iter__(self):
        return self
    
    def __next__(self):
        while self.index < len(self.data):
            item = self.data[self.index]
            self.index += 1

            if item not in self.seen:
                self.seen.add(item)
                return item
        raise StopIteration


data = [1, 2, 2, 3, 4, 3, 5, 1]

for value in items(data):
    print(value)  
```

14. Implement a decorator to check if inputs of a function are positive numbers.

```python
def check_positive(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        for value in args:
            if not isinstance(value, (int, float)):
                raise TypeError("All inputs must be numbers")
            if value <= 0:
                raise ValueError("All inputs must be positive numbers")
        return func(*args, **kwargs)
    return wrapper

@check_positive
def multiply(a, b):
    return a * b

print(multiply(3, 4))    

# print(multiply(-3, 4))   
```

15. Write a generator to simulate an infinite sequence of random numbers.

```python
import random

def random_number_stream():
    while True:
        yield random.random()   # random float between 0 and 1
rng = random_number_stream()

for _ in range(5):   
    print(next(rng))
```
