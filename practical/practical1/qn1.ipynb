1.Write a Python program to check if a given number is prime.

```python
num = int(input('Enter an integer number: '))

if num <= 1:
    print("Not a prime number")
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

```python
# 2. Create a function that converts a list of temperatures from Celsius to Fahrenheit.

def convert(temp):
    fahrenheit_list=[]
    for c in temp:
        f=(c* 9/5)+32
        fahrenheit_list.append(f)
    return fahrenheit_list
temp_c=[0,34,65]
temp_f=convert(temp_c)
print(temp_f)
```

```python
#4. Implement a function to find the factorial of a number using recursion
def fact(num):
    if num==0 or num==1:
        return 1
    else:
        return num*fact(num-1)

n=int(input('Enter a number: '))
print('Factorial:',fact(n))
```

```python
# 6. Create a function that takes a list of numbers and returns only the even numbers.
def even(num):
    even=[]
    for n in num:
        if n%2==0:
            even.append(n)
    return even

nums=list(map(int,input("Enter an Integer : ").split()))
print(even(nums))
```

```python
# 8. Implement a class Circle with a method to calculate the area and circumference.
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def circumference(self):
        return 2 * math.pi * self.radius
r = float(input("Enter radius: "))
c = Circle(r)

print("Area:", c.area())
print("Circumference:", c.circumference())
```

```python
# 11. Implement a function to remove duplicates from a list without using set().
def remove_duplicates(lst):
    result=[]
    for item in lst:
        if item not in result:
            result.append(item)
    return result

nums=[1,2,3,4,5,6,5,6]
print(remove_duplicates(nums))
```

```python
# 12. Write a Python program to find the largest and smallest numbers in a list.
def find_largest_smallest(lst):
    largest = lst[0]
    smallest = lst[0]

    for item in lst:
        if item > largest:
            largest = item
        if item < smallest:
            smallest = item

    return largest, smallest

numbers = [20, 5, 7, 25, 2]
large, small = find_largest_smallest(numbers)

print("Largest:", large)
print("Smallest:", small)

```

```python
# 13. Create a class Rectangle with methods to calculate area and perimeter.
class Rectangle:
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2*(self.length+self.width)

rect=Rectangle(20,10)

print("Area of Rectangle is : ",rect.area())
print("Perimeter of Rectangle is : ",rect.perimeter())
```

```python
# 14. Write a function to reverse a list without using the built-in reverse() method.
def reverse_lst(lst):
    reversed_lst=[]
    for item in lst:
        reversed_lst.insert(0,item)
    return reversed_lst

numbers=[1,2,3,4,5]
print(reverse_lst(numbers))
    
```

```python
# 15. Implement a Python function that generates the Fibonacci sequence up to n terms
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

terms = int(input('enter the terms number: '))
for i in range(terms):
    print(fibonacci(i), end=" ")


```

```python
# 16. Write a program to count the frequency of each word in a given string.
def word_frequency(text):
    words=text.split()
    freq={}

    for word in words:
        if word in freq:
            freq[word]+=1
        else:
            freq[word]=1
    return freq

sentence="Hello python I am Advanced python for data science"
result=word_frequency(sentence)

for word,count in result.items():
    print(word, ":",count)
```

```python
# 17. Create a function that accepts a variable number of arguments and returns their sum.
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_numbers(1, 2, 3, 4, 5))
```

```python
# 18. Implement a class Employee with attributes name and salary and a method to
# apply a 10% raise.
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

    def apply_raise(self):
        self.salary=self.salary*1.10 

    def display(self):
        print("Name:",self.name)
        print("Salary:",self.salary)

emp=Employee("susan",20000)
emp.apply_raise()
emp.display()
```

```python
# 19. Write a Python program to check if two strings are anagrams of each other.
def is_anagram(s1, s2):
    s1 = s1.replace(" ", "").lower()
    s2 = s2.replace(" ", "").lower()

    return sorted(s1) == sorted(s2)


print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
```

```python
# 20. Create a function that calculates the sum of digits of a given number.
def sum_of_digits(num):
    total = 0
    num = abs(num)   

    while num > 0:
        digit = num % 10
        total += digit
        num //= 10

    return total


print(sum_of_digits(1234))   
```
