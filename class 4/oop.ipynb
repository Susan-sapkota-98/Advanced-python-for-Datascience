```python
class Animal:
    def __init__(self,name):
        self.name=name

    def speak(self):
        print(f"{self.name} SPEAKS !")

class Dog(Animal):
    def __init__(self,name,breed):
        super().__init__(name)  #super() is a function
        self.breed=breed

    def speak(self):
        super().speak()
        print(f"{self.name} BARKS !")

d1=Dog("Lucy","Shepherd")
d1.speak()
d2=Animal("Lucky")
d2.speak()
        
```

```python
class A:
    def show(self):
        print("A")

class B:
    def show(self):
        print("B")
class C(A,B):
    pass
c1=C()
c1.show()
```

```python
#method resolution order
C.mro()
```
